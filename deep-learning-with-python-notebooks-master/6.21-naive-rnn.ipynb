```python
import numpy as np
```

```python
timesteps = 10        # 100
input_features = 3    # 32
output_features = 6   # 64
```

```python
inputs = np.random.random((timesteps, input_features))
print(type(inputs))
print(inputs)
```

```python
state_t = np.zeros((output_features, ))
print(type(state_t))
print(state_t)
```

```python
W = np.random.random((output_features, input_features))
print(type(W))
print(W)
```

```python
U = np.random.random((output_features, output_features))
print(type(U))
print(U)
```

```python
b = np.random.random((output_features, ))
print(b)
```

```python
successive_outputs = []
print(type(successive_outputs))
print(successive_outputs)
```

```python
print(W)
print(inputs[0])
np.dot(W, inputs[0])
```

```python
print(U)
print(state_t)
np.dot(U, state_t)
```

```python
print(inputs)
for input_t in inputs:
    print('--------------------------------------------------------------')
    print('input_t=' + str(input_t))
    print('state=' + str(state_t))
    print('~~~~~~~~~')
    output_t = np.tanh(np.dot(W, input_t) + np.dot(U, state_t) + b)
    print('np.dot(W, input_t):')
    print(np.dot(W, input_t))
    print('~~~~~~~~~')
    print('np.dot(U, state_t):')
    print(np.dot(U, state_t))
    print('~~~~~~~~~')
    print('output_t=' + str(output_t))
    print('~~~~~~~~~')
    successive_outputs.append(output_t)
    print('successive_outputs' + str(successive_outputs))
    
    state_t = output_t
   
```

```python
final_output_sequence = np.concatenate(successive_outputs, axis=0)
print(final_output_sequence)
```
